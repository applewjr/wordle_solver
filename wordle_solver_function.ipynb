```python
def wordle_solver(import_path: str, must_be_present: str, must_not_be_present: str, present: str,
    value1_not_present: list, value2_not_present: list, value3_not_present: list, value4_not_present: list, value5_not_present: list):

    must_be_present = list(must_be_present)
    must_not_be_present = list(must_not_be_present)
    present = list(present)
    # not_present = list(not_present)
    
    places = ['one', 'two', 'three', 'four', 'five']
    df = pd.read_excel(import_path)
    total_len = len(df)

    # must be present
    for j in must_be_present:
        drop_list = []
        for i in range(len(df)):
            drop_list.append(df['word'][i].find(j))
        df['drop_no_' + j] = drop_list
    for j in must_be_present:
        df = df[df['drop_no_' + j] != -1]

    # must not be present
    for i in places:
        for j in must_not_be_present:
            df = df[df[i] != j]

    # specific values must be present
    if present[0] != ' ':
        df = df[df['one'] == present[0]]
    if present[1] != ' ':
        df = df[df['two'] == present[1]]
    if present[2] != ' ':
        df = df[df['three'] == present[2]]
    if present[3] != ' ':
        df = df[df['four'] == present[3]]
    if present[4] != ' ':
        df = df[df['five'] == present[4]]


    # specific values not must be present
    if len(value1_not_present) > 0:
        for i in value1_not_present:
            df = df[df['one'] != (','.join(i))]
    if len(value2_not_present) > 0:
        for i in value2_not_present:
            df = df[df['two'] != (','.join(i))]
    if len(value3_not_present) > 0:
        for i in value3_not_present:
            df = df[df['three'] != (','.join(i))]
    if len(value4_not_present) > 0:
        for i in value4_not_present:
            df = df[df['four'] != (','.join(i))]
    if len(value5_not_present) > 0:
        for i in value5_not_present:
            df = df[df['five'] != (','.join(i))]

    # pick the best (aka reasonably good) choice by sorting on the highest 'best_word'
    df = df.sort_values(by = 'best_word', ascending =  False)

    try:
        print(df['word'].iloc[0:5])
    except:
        print('no word found')
    
    print(f'Options remaining: {len(df)}/{total_len} ({round(len(df)/total_len*100,2)}%)')
```
